### Get counts and barplot

```python
df = pd.read_csv(r'C:\Users\30694\Desktop\ML-STATS\ML-AI\Theo_ML\PANDAS\pandas-cookbook-master\data\311-service-requests.csv')
df[:3]
```

```python
complaint_counts = df['Complaint Type'].value_counts()
complaint_counts
```

```python
complaint_counts[:10].plot(kind='bar')
```

### Barplot on a conditioned series object

```python
is_noise = df['Complaint Type'] == "Noise - Street/Sidewalk"
noise_complaints = df[is_noise]
noise_complaints['Borough'].value_counts()
```

```python
noise_complaint_counts = noise_complaints['Borough'].value_counts()
complaint_counts = df['Borough'].value_counts()
noise_complaint_counts / complaint_counts
```

```python
(noise_complaint_counts / complaint_counts.astype(float)).plot(kind='bar')
```
