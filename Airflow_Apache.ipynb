Apache Airflow is not a DevOps tool. It is a workflow orchestration tool primarily designed for managing “ETL” jobs in Hadoop environments. It basically will execute commands on the specified platform and also orchestrate data movement. It was never designed to do anything remotely similar to Jenkins or Gitlab.

It is based on a framework that was created at Facebook called Dataswarm. The creator of Airflow worked at Facebook (we worked on the same team at one point) and when he moved to AirBNB, he took the concepts and created an open source framework. It has gained a ton of popularity because for the types of workloads it orchestrates, it does a very good job. And Apache Oozie (which has been part of the Hadoop stack for a long time) is really quite bad at doing the same things.
