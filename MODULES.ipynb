* finders
* loaders
* finders + loader == importer

find --> load --> compile --> execute --> put into cache  --> put symbol in our namespace that points to the compiled obj in memory

finders return a ModuleSpec

```python
import fractions
fractions.__spec__
```

```python
# sys.meta_path contains the finders
sys.meta_path
```

```python
# BuiltinImporter is a finder and loader
```

when you do import --> find asks finders if they know where to find the module --> if it finds it it return the module spec  --> then the loader, loads the module and then compile, exec put into sys.modules.

* You can write your own finders and loaders to load a module from a database or from a REST API call!
